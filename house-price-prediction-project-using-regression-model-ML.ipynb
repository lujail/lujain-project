```python
import pandas as pd
data = pd.read_csv("/kaggle/input/house-prices-advanced-regression-techniques/train.csv")
```

```python
X = data.drop(columns=["SalePrice"])  

y = data["SalePrice"]
```

```python
print(X.head())
print(X.columns)
```

```python
X = pd.get_dummies(X, drop_first=True) 
X = X.fillna(X.mean())
```

```python
print(X.select_dtypes(include=['object']).head())
```

```python
X = X.astype({col: 'int64' for col in X.select_dtypes(include=['bool']).columns})
```

```python
print(X.isnull().sum())


X = X.fillna(X.mean())
```

```python
print(X.dtypes)


non_numeric_cols = X.select_dtypes(exclude=['number']).columns
if non_numeric_cols.empty:
    print("All columns are numeric!")
else:
    print("Non-numeric columns still exist:", non_numeric_cols)
```

```python
import numpy as np

print("NaN values:", np.isnan(X.values).sum())  
print("Infinite values:", np.isinf(X.values).sum()) 
```

```python
from sklearn.feature_selection import VarianceThreshold


selector = VarianceThreshold(threshold=0.01) 
X_reduced = selector.fit_transform(X)


X = pd.DataFrame(X_reduced, columns=X.columns[selector.get_support()])
```

```python
correlation_matrix = X.corr()


perfectly_correlated = correlation_matrix[correlation_matrix == 1.0].stack().reset_index()
perfectly_correlated = perfectly_correlated[perfectly_correlated['level_0'] != perfectly_correlated['level_1']]
print("Perfectly correlated features:\n", perfectly_correlated)
```

```python
from sklearn.feature_selection import VarianceThreshold


selector = VarianceThreshold(threshold=0.01) 


selector.fit(X)


X = X.iloc[:, selector.get_support(indices=True)]
```

```python
print(X.dtypes)  
print(X.isnull().sum())  
```

```python
remaining_features = X.columns
print("Remaining features:", remaining_features)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd


vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif_data) 
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.01)
lasso.fit(X, y)
selected_features = X.columns[(lasso.coef_ != 0)]
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

model = LinearRegression()
rfe = RFE(model, n_features_to_select=10)  
X_rfe = rfe.fit_transform(X, y)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import train_test_split 
```

```python
X_train = vif_data  
X_test = test_data
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Random Forest Classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
rf_preds = rf.predict(X_test)
```

```python
# Gradient Boosting Classifier
gb = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
gb.fit(X_train, y_train)
gb_preds = gb.predict(X_test)
```

```python
import pandas as pd
```

```python
test_data = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/test.csv')
```

```python
if len(rf_preds) < len(test_data):
    mean_rf = np.mean(rf_preds)  
    rf_preds = np.pad(rf_preds, (0, len(test_data) - len(rf_preds)), constant_values=mean_rf) 

if len(gb_preds) < len(test_data):
    mean_gb = np.mean(gb_preds)  
    gb_preds = np.pad(gb_preds, (0, len(test_data) - len(gb_preds)), constant_values=mean_gb)  
```

```python
submission_rf = pd.DataFrame({'Id': test_data['Id'],'SalePrice': rf_preds })
submission_rf.to_csv('submission_rf.csv', index=False)
print(submission_rf) 
```

```python
submission_gb = pd.DataFrame({'Id': test_data['Id'], 'SalePrice': gb_preds })
submission_gb.to_csv('submission_gb.csv', index=False) 
print(submission_gb)
```
